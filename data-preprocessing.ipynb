```python
shop_in = './Data/shopping_data.csv'

shop_df = pd.read_csv(shop_in)

shop_df
```

```python
print(shop_df.columns)

print(shop_df.dtypes)
```

```python
# loop through the columns to identify missing values

for column in shop_df.columns:
    print(f'Column {column} has {shop_df[column].isnull().sum()} null values')
    
```

```python
# locate any duplicate data entries

print(f'Duplicate entries: {shop_df.duplicated().sum()}')
```

```python
# drop the CustomerID column

shop_df.drop(['CustomerID'], axis=1, inplace=True)
shop_df
```

```python
# Transform the data

def string_value(member):
    if member == 'Yes':
        return 1
    else:
        return 0
shop_df['Card Member'] = shop_df['Card Member'].apply(string_value)

shop_df
```

```python
# Scale the annual income values

shop_df['Annual Income'] = shop_df['Annual Income'] /1000

shop_df
```

```python
shop_df = shop_df.rename(columns={'Card Member': 'card_member', 'Age': 'age', 'Annual Income': 'annual_income', 'Spending Score (1-100)': 'spending_score'})

shop_df = shop_df.dropna(axis=0, how='any')
shop_df
```

```python
# Save the cleaned data to csv

shop_outfile = './Data/shopping_cleaned_data.csv'

shop_df.to_csv(shop_outfile, index=False)

shop_xl = './Data/shopping_cleaned_data.xlsx'

shop_df.to_excel(shop_xl, index=False)
```
